### Ejercicio 1

```python
# Corpus simple para practicar
sinopsis_ejemplo = [
    "Una historia de amor en tiempos de guerra",
    "Aventuras mágicas en un mundo fantástico",
    "Un detective resuelve misterios en la ciudad",
    "Romance juvenil en una escuela",
    "Ciencia ficción sobre viajes en el tiempo"
]

def ejercicio_vectorizacion_basica(sinopsis):
    """
    Implementar y comparar:
    1. Count Vectorizer básico
    2. TF-IDF básico
    3. Mostrar las matrices resultantes
    4. Calcular similitud entre la primera y segunda sinopsis
    """
    print("=== EJERCICIO DE VECTORIZACIÓN BÁSICA ===\n")

    # 1. Count Vectorizer básico
    print("1. COUNT VECTORIZER")
    print("-" * 40)

    count_vectorizer = CountVectorizer()
    count_matrix = count_vectorizer.fit_transform(sinopsis)

    print("Vocabulario:", count_vectorizer.get_feature_names_out())
    print("\nMatriz Count Vectorizer:")
    print(count_matrix.toarray())

    # Mostrar como DataFrame para mejor visualización
    count_df = pd.DataFrame(count_matrix.toarray(),
                           columns=count_vectorizer.get_feature_names_out(),
                           index=[f"Doc {i+1}" for i in range(len(sinopsis))])
    print("\nMatriz como DataFrame:")
    print(count_df)

    # 2. TF-IDF básico
    print("\n\n2. TF-IDF VECTORIZER")
    print("-" * 40)

    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(sinopsis)

    print("Vocabulario:", tfidf_vectorizer.get_feature_names_out())
    print("\nMatriz TF-IDF:")
    print(tfidf_matrix.toarray())

    # Mostrar como DataFrame para mejor visualización
    tfidf_df = pd.DataFrame(tfidf_matrix.toarray(),
                           columns=tfidf_vectorizer.get_feature_names_out(),
                           index=[f"Doc {i+1}" for i in range(len(sinopsis))])
    print("\nMatriz TF-IDF como DataFrame:")
    print(tfidf_df.round(3))

    # 3. Calcular similitud entre la primera y segunda sinopsis
    print("\n\n3. CÁLCULO DE SIMILITUD")
    print("-" * 40)

    # Similitud usando Count Vectorizer
    count_sim = cosine_similarity(count_matrix[0:1], count_matrix[1:2])[0][0]
    print(f"Similitud coseno (Count Vectorizer) entre Doc1 y Doc2: {count_sim:.4f}")

    # Similitud usando TF-IDF
    tfidf_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
    print(f"Similitud coseno (TF-IDF) entre Doc1 y Doc2: {tfidf_sim:.4f}")

    # 4. Comparación de todas las similitudes
    print("\n\n4. MATRIZ DE SIMILITUDES COMPLETA")
    print("-" * 40)

    # Matriz de similitudes Count Vectorizer
    count_similarities = cosine_similarity(count_matrix)
    print("Similitudes Count Vectorizer:")
    count_sim_df = pd.DataFrame(count_similarities,
                               index=[f"Doc {i+1}" for i in range(len(sinopsis))],
                               columns=[f"Doc {i+1}" for i in range(len(sinopsis))])
    print(count_sim_df.round(3))

    # Matriz de similitudes TF-IDF
    tfidf_similarities = cosine_similarity(tfidf_matrix)
    print("\nSimilitudes TF-IDF:")
    tfidf_sim_df = pd.DataFrame(tfidf_similarities,
                               index=[f"Doc {i+1}" for i in range(len(sinopsis))],
                               columns=[f"Doc {i+1}" for i in range(len(sinopsis))])
    print(tfidf_sim_df.round(3))

    return count_matrix, tfidf_matrix, count_similarities, tfidf_similarities

# Ejecutar el ejercicio
count_matrix, tfidf_matrix, count_sim, tfidf_sim = ejercicio_vectorizacion_basica(sinopsis_ejemplo)
```

```python
from typing import Any

class Clave():
  """
  Almacena la clave real a ser hasheada,
  usando la representación que se desee,
  y encapsula el algoritmo de hash elegido.
  """

  def __init__(self, key: str):
    self.key = key

  def to_int(self) -> int:
    """
    Esta función debe convertir el valor de la key en un número.
    """
    key_number=hash(self.key) # Lo que hace la funcion hash es crear un numero entero aleatorio dependiendo la cadena
    ascii_number=sum(ord(caracter) for caracter in self.key) # Aca estamos usando la funcion ord que lo que hace es devolver el valor entero del caracter segun la tabla ascii
    return key_number

  def __eq__(self, other) -> bool:
    """
    Decide si dos claves son iguales
    """
    return self.key == other.key
```

```python
from typing import Any

class Nodo:
    def __init__(self, dato: Any, prox: Any = None) -> None:
        self.dato = dato
        self.prox = prox

    def __str__(self):
        return str(self.dato)

    def view_list(self, nodo: Any = None) -> None:
        while nodo is not None:
            print(nodo)
            nodo = nodo.prox
```

```python
class ListaEnlazada(Nodo):
  def __init__(self):
    """Esta lista esta implementada para recibirla vacia"""
    self.primero = None
    self.len = 0

  def __len__(self):
    return int(self.len)

  def insert(self,i,x):
    """
    Insertamos elementos a la lista enlazada vacia.
    """
    if i < 0 or i>self.len:
      print("Valor invalido")
      return
    nuevo = Nodo(x)
    if i == 0:
      nuevo.proximo = self.primero
      self.primero = nuevo
    else:
      n_ant = self.primero
      for pos in range(1,i):
        n_ant= n_ant.proximo
      nuevo.proximo = n_ant.proximo
      n_ant.proximo = nuevo
    self.len += 1

  def pop(self,i = None):
    """
    Esta funcion se utilizara para eliminar un valor de la lista por indice
    Si no recibe ningun indice se tomara como el ultimo valor de la lista(self.len-1 ya que el ultimo al apuntar siempre es None)
    """
    if i == None:
      i = self.len -1
    if i < 0 or i > self.len:
      print("Posicion invalida")
      return
    if i == 0:
      dato = self.primero.dato
      self.primero = self.primero.proximo
      return dato
    n_ant = self.primero
    n_act = n_ant.proximo
    for pos in range(1,i):
      n_ant = n_act
      n_act = n_ant.proximo
    """
    n_ant annterior al buscado, n_act -> nodo que buscamos eliminar
    """
    dato = n_act.dato
    n_ant.proximo = n_act.proximo
    self.len -= 1
    return dato

  def remove(self, x):
    if self.primero is None:
        print("Lista vacia.")
        return

    if self.primero.dato[0] == x:
        self.primero = self.primero.proximo
        self.len -= 1
        return

    n_ant = self.primero
    n_act = n_ant.proximo

    while n_act is not None and n_act.dato[0] != x:
        n_ant = n_act
        n_act = n_act.proximo

    if n_act is None:
        print("El valor no se encuentra")
        return

    n_ant.proximo = n_act.proximo
    self.len -= 1

  def append(self, x):
    '''
    Esta funcion lo que hace es agregar un elemento al final de la lista.
    '''
    nuevo_nodo=Nodo(x)
    if not self.primero:
      self.primero=nuevo_nodo
    else:
      actual=self.primero
    while actual.proximo:
      actual=actual.proximo
    actual.proximo=nuevo_nodo
    self.len+=1

  def index(self, x):
    '''
    El index devuelve la primera aparicion de X que seria un valor ingresado
    '''
    actual=self.primero
    indice=0
    while actual is not None:
      if actual.dato==x:
        return indice
      actual=actual.proximo
      indice+=1
    raise ValueError('No esta en la lista')
```

```python
# No es necesario cambiar nada en esta celda! Esta celda solo define la interfaz
# del TAD Diccionario.
from typing import Any

class Diccionario():
  """
  Interfaz del TAD Diccionario
  - insert(key, value) - Inserta un elemento con clave key y valor value en el diccionario.
      Si la clave ya se encuentra en el diccionario, debe reemplazar el value anterior por el nuevo.
  - search(key) - Devuelve la el value asociado con la clave key, o muestra un mensaje de error si la clave no se encuentra.
  - delete(key) - Elimina el elemento con clave key del diccionario.
  """

  def insert(self, key: Clave, data: Any) -> None:
    pass

  def search(self, key: Clave) -> Any:
    pass

  def delete(self, key: Clave) -> None:
    pass
```

```python
class HashTable(Diccionario):
  """
  Implementación de Diccionario con tabla hash
  de tamaño fijo y resolución de colisiones
  por encadenamiento
  """

  def __init__(self, size: int=0):
    """
    El constructor inicializa una tabla de tamaño `size`.
    """
    super().__init__()
    self.size = size
    self.table = [None] * size

  def insert(self, key: Clave, data: Any) -> None:
    """
    Implementacion del metodo insert
    """
    position=key.to_int() % self.size

    if self.table[position] is None:
      self.table[position]=ListaEnlazada()

    self.table[position].insert(0, [key.key,data])



  def search(self, key: Clave) -> Any | None:
    """
    Implementacion del metodo Search.
    """
    position=key.to_int() % self.size
    lista_enlazada=self.table[position]

    if lista_enlazada is not None:
      valor_actual=lista_enlazada.primero
      while valor_actual is not None:
        if valor_actual.dato[0]==key.key:
          return valor_actual.dato[1]
        valor_actual=valor_actual.prox
    return None


  def delete(self, key: Clave):
    """
    Implementacion del metodo Delete.
    """
    position = key.to_int() % self.size
    lista_enlazada = self.table[position]

    if lista_enlazada is not None:
        lista_enlazada.remove(key.key)

    print(f"Key not found: {key.key}")

  def __len__(self) -> int:
    """
    Este método sobrecarga la función `len`.
    Debe devolver la cantidad de elementos insertados en la tabla.
    Si la tabla esta vacía, devuelve cero.
    """
    return sum(len(lista) if lista is not None else 0 for lista in self.table)

  def _load_factor(self) -> float:
    """
    Este método debe devolver el factor de carga de la tabla
    """
    return len(self)/self.size if self.size > 0 else 0.0

  def _longest_list(self) -> int:
    """
    Este método debe devolver la longitud de la lista mas larga de la tabla
    """
    return max(len(lista) if lista is not None else 0 for lista in self.table)

  def __str__(self) -> str:
    """
    Mostrar la cantidad de elementos de la tabla (__len__), el factor de carga (_load_factor)
    y la longitud de la lista mas larga (_longest_list)
    """
    return f"Elementos: {len(self)}, Factor de Carga: {self._load_factor()}, Longitud Lista Más Larga: {self._longest_list()}"
```
